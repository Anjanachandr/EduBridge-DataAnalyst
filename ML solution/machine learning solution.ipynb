```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=21)
```

```python
for i in range(0, len(clmn2 )):
    sns.boxplot(y=df[clmn2[i]],color='blue',orient='v')
    plt.show()
```

```python
for i in range(len(clmn2)): 
    sns.boxplot(df[clmn2[i]],color='blue',orient='v')
    plt.title(clmn2[i])
    plt.title("Before removing outlier")
    plt.show()
    def drop_outliers(df, field_name):
        iqr = 1.5 * (np.percentile(df[field_name], 75) - np.percentile(df[field_name], 25))
        df.drop(df[df[field_name] > (iqr + np.percentile(df[field_name], 75))].index, inplace=True)
        df.drop(df[df[field_name] < (np.percentile(df[field_name], 25) - iqr)].index, inplace=True)
        iqr = 1.5 * (np.percentile(df[field_name], 75) - np.percentile(df[field_name], 25))
        df.drop(df[df[field_name] > (iqr + np.percentile(df[field_name], 75))].index, inplace=True)
        df.drop(df[df[field_name] < (np.percentile(df[field_name], 25) - iqr)].index, inplace=True)
    drop_outliers(df, clmn2[i])
    sns.boxplot(df[clmn2[i]],color='blue',orient='v')
    plt.title("After removing outlier")
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
SC=StandardScaler()
x_train=SC.fit_transform(x_train)
x_test=SC.fit_transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier  
classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
parameters ={ 'n_neighbors' : [5,7,9,11,13,15],
               'weights' : ['uniform','distance'],
               'metric' : ['minkowski','euclidean','manhattan']}
from sklearn.model_selection import RandomizedSearchCV
cv = RandomizedSearchCV(classifier,parameters ,cv=5) 
```

```python
cv.fit(x_train,y_train)
```

```python
y_pred = cv.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print('\n knn accuracy:',accuracy_score(y_pred,y_test))
```
